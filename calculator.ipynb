```python
from tkinter import*

def clear():
    entry.delete(0,END)
    
def clear_c():
    a=entry.get()
    length=len(a)
    entry.delete(length-1,END)
    
def fnctn():
    global k
    b=str(entry.get())
    c=eval(b)
    k=('=',c)
    return k


def equal():
    try:
        fnctn()
        entry.insert(END,k)
    except:
        entry.delete(0, END)
        entry.insert(0,'SYNTAX ERROR')



    
    
    
window=Tk()
window.title('CALCULATOR')
window.geometry('300x300')


entry=Entry(window,width=32)



bt0=Button(window,text=0,width=5,command=lambda : entry.insert(END, 0))
bt1=Button(window,text=1,width=5,command=lambda : entry.insert(END, 1))
bt2=Button(window,text=2,width=5,command=lambda : entry.insert(END, 2))
bt3=Button(window,text=3,width=5,command=lambda : entry.insert(END, 3))
bt4=Button(window,text=4,width=5,command=lambda : entry.insert(END, 4))
bt5=Button(window,text=5,width=5,command=lambda : entry.insert(END, 5))
bt6=Button(window,text=6,width=5,command=lambda : entry.insert(END, 6))
bt7=Button(window,text=7,width=5,command=lambda : entry.insert(END, 7))
bt8=Button(window,text=8,width=5,command=lambda : entry.insert(END, 8))
bt9=Button(window,text=9,width=5,command=lambda : entry.insert(END, 9))
bt10=Button(window,text='+',width=5,command=lambda : entry.insert(END, '+'))
bt11=Button(window,text='-',width=5,command=lambda : entry.insert(END, '-'))
bt12=Button(window,text='x',width=5,command=lambda : entry.insert(END, '*'))
bt13=Button(window,text='/',width=5,command=lambda : entry.insert(END, '/'))
bt14=Button(window,text='=',width=5,command=equal)
bt15=Button(window,text='.',width=5,command=lambda : entry.insert(END, '.'))
bt16=Button(window,text='AC',command=clear,width=5,bg='green')
bt17=Button(window,text='(',width=5,command=lambda : entry.insert(END, '('))
bt18=Button(window,text=')',width=5,command=lambda : entry.insert(END, ')'))
bt19=Button(window,text='%',width=5,command=lambda : entry.insert(END, '/100'))
bt20=Button(window,text='C',command=clear_c,width=5)

bt7.place(x=50,y=100)
bt8.place(x=100,y=100)
bt9.place(x=150,y=100)
bt4.place(x=50,y=130)
bt5.place(x=100,y=130)
bt6.place(x=150,y=130)
bt1.place(x=50,y=160)
bt2.place(x=100,y=160)
bt3.place(x=150,y=160)
bt0.place(x=100,y=190)
bt15.place(x=150,y=190)
bt11.place(x=200,y=130)
bt12.place(x=200,y=100)
bt13.place(x=50,y=190)
bt20.place(x=200,y=190)
bt10.place(x=200,y=160)
bt16.place(x=200,y=70)
bt19.place(x=50,y=70)

bt17.place(x=100,y=70)
bt18.place(x=150,y=70)
bt14.place(x=200,y=220)

entry.place(x=50,y=40)




















window.mainloop()
```
